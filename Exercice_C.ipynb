```python
import numpy as np
import pandas as pd
import math
```

```python
df=pd.read_csv("Natixis.csv", sep=";")
df["Cours de Natixis"] = df["Cours de Natixis"].astype(str).str.replace(",", ".").astype(float)
```

```python
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
P = df["Cours de Natixis"].values 
returns = (P[1:] - P[:-1]) / P[:-1]
returns = np.append(returns, np.nan)

df["Return"] = returns
```

```python
returns_sorted = np.sort(df["Return"].values)


losses = returns_sorted[returns_sorted < 0]
gains = returns_sorted[returns_sorted > 0]

abs_losses = np.abs(losses)
abs_losses = np.sort(abs_losses)
```

```python
abs_losses = np.abs(losses)
abs_losses = np.sort(abs_losses)
```

We calculate Pickand estimator for losses using the formula

```python
n = len(losses)
k = int(np.sqrt(n))
pickands_losses = (1 / np.log(2)) * np.log((abs_losses[-k] - abs_losses[-2*k]) / (abs_losses[-2*k] - abs_losses[-4*k]))
```

```python
print(pickands_losses)
```

We calculate Pickand estimator for losses using the formula

```python
n_gain=len(gains)
k_gain=int(np.sqrt(n_gain))
pickands_gains=(1 / np.log(2)) * np.log((gains[-k_gain] - gains[-2*k_gain]) / (gains[-2*k_gain] - gains[-4*k_gain]))
```

```python
print(pickands_gains)
```

Now using the formula of the pickand VaR with the pickand estimator with iid assumptions 

```python
alpha = 0.95

VaR_95 = ((((k / (n * (1 - alpha))) ** pickands_losses**alpha) - 1)/ (1 - 2 ** (-pickands_losses*alpha))) * (abs_losses[-k] - abs_losses[-2 * k]) + abs_losses[-k]
```

```python
print(VaR_95)
```
