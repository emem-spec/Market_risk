```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Dataset TD4.xlsx")
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df['price_change'] = df['Price (before transaction)'].diff()
```

```python
df.head()
```

```python
known_value=df.dropna(subset=["volume of the transaction (if known)"])
```

```python
X_known = known_value[['price_change', 'bid-ask spread', 'Sign of the transaction']]
y_known = known_value['volume of the transaction (if known)']
```

```python
X_known = X_known.fillna(0)
model = LinearRegression()
model.fit(X_known, y_known)
```

```python
unknown=df[df["volume of the transaction (if known)"].isna()]
```

```python
X_unknown = unknown[['price_change', 'bid-ask spread', 'Sign of the transaction']].fillna(0)
predicted_volumes = model.predict(X_unknown)

df.loc[df['volume of the transaction (if known)'].isna(), 'estimated_volume'] = predicted_volumes
print(df[['transaction date (1=1day=24 hours)', 'volume of the transaction (if known)', 'estimated_volume']].head(10))
```

```python
df["volume of the transaction (if known)"]=df["volume of the transaction (if known)"].combine_first(df["estimated_volume"])
```

```python
df.head()
```

```python
df.drop(columns="estimated_volume")
```

```python
df = df.dropna(subset=["price_change"])

df["y"] = df["price_change"] / df["Sign of the transaction"]

X = df[["volume of the transaction (if known)"]]
y = df["y"]

# Fit linear regression
model = LinearRegression()
model.fit(X, y)

gamma = model.coef_[0]   # slope
etha = model.intercept_   # intercept

print(f"Estimated gamma (permanent impact): {gamma:.6f}")
print(f"Estimated etha (temporary impact): {etha:.6f}")
```
