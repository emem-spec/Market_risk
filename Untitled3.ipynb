```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Dataset TD5.xlsx")
```

```python
df.head()
```

```python
new_header = df.iloc[1]
df=df[2:]
df.columns=new_header
df = df.reset_index(drop=True)
df.head()
```

```python
GBP=df.iloc[:,:3]
```

```python
GBP.head()
```

```python
SEK=df.iloc[:,4:7]
SEK.head()
```

```python
CAD=df.iloc[:,8:11]
CAD.head()
```

```python
GBP["average"]=(GBP["HIGH"]+GBP["LOW"])/2
SEK["average"]=(SEK["HIGH"]+SEK["LOW"])/2
CAD["average"]=(CAD["HIGH"]+CAD["LOW"])/2
```

```python
GBP["t"]=GBP.index
SEK["t"]=SEK.index
CAD["t"]=CAD.index
```

```python
GBP["k"]=GBP["t"]-1
SEK["k"]=SEK["t"]-1
CAD["k"]=CAD["t"]-1
```

```python
GBP.head()
```

```python
j=-7
GBP["k2^-j"]=GBP["k"]*2**(-j)
SEK["k2^-j"]=SEK["k"]*2**(-j)
CAD["k2^-j"]=CAD["k"]*2**(-j)
```

```python
GBP["(k+1)2^-j"]=(GBP["k"]+1)*2**(-j)
SEK["(k+1)2^-j"]=(SEK["k"]+1)*2**(-j)
CAD["(k+1)2^-j"]=(CAD["k"]+1)*2**(-j)
```

```python
resultats = []
some_locale=0
for k in GBP["k"]:
    borne_inf = k * 2**(-j)        
    borne_sup = (k + 1) * 2**(-j)  
    
    # C'est l'équivalent de [indicatrice(A) - indicatrice(B)]
    masque = (GBP["t"] >= borne_inf) & (GBP["t"] < borne_sup)
    
    somme_locale = np.sum(GBP["average"][masque])
    resultats.append(2**(j/2) * somme_locale)

GBP["coef wavelet"] = resultats
```
