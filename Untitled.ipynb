```python
import numpy as np
import pandas as pd
import math
```

```python
df=pd.read_csv("TD1.csv", sep=";")
df["Cours de Natixis"] = df["Cours de Natixis"].astype(str).str.replace(",", ".").astype(float)
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
P = df["Cours de Natixis"].values 
returns = (P[1:] - P[:-1]) / P[:-1]
returns = np.append(returns, np.nan)

df["Return"] = returns

df_date = df[(df["Date"] >= "2015-01-01") & (df["Date"] <= "2016-12-31")] 

df_sorted = df_date.sort_values(by = "Return", ascending = True)

df_date.head()

```

```python
alpha = 0.05  # probability level (5%)
returns = df_date["Return"]

returns_sorted = np.sort(returns)
n=len(returns_sorted)
# Find the index corresponding to alpha
index_empirical = int(np.floor(alpha * (n-1)))

# Empirical VaR
VaR_empirical = returns_sorted[index_empirical]

print(f"Empirical VaR at {alpha*100}%: {VaR_empirical:}")
```

```python
from scipy.stats import norm

mu = returns.mean()
sigma = returns.std()
z = norm.ppf(alpha)

VaR_parametric = mu + sigma * z
print(f"Parametric Gaussian VaR at {alpha*100}%: {VaR_parametric:.4f}")
```

```python
from scipy.stats import gaussian_kde

kde = gaussian_kde(returns)

x_grid = np.linspace(returns.min(), returns.max(), 10000)
cdf = np.cumsum(kde(x_grid))
cdf /= cdf[-1]  

VaR_kernel = x_grid[np.searchsorted(cdf, alpha)]
print(f"Non-parametric KDE VaR at {alpha*100}%: {VaR_kernel:.4f}")
```

```python
methods = {
    "Empirical": VaR_empirical,
    "Parametric": VaR_parametric,
    "Kernel": VaR_kernel
}
n=len(returns)
print(n)
pow(n,3)
```

```python
results = []
for name, VaR in methods.items():
    exceptions = int(np.sum(returns <= VaR))
    proportion = exceptions / n
    results.append([name, exceptions, proportion])

table = pd.DataFrame(results, columns=["Method", "Exceptions", "Proportion"])
print(table.to_string(index=False))
```

```python
# Filter returns for October â€“ December 2016
df_date3 = df[(df["Date"] >= "2016-10-01") & (df["Date"] <= "2016-12-31")]
df_date3 = df_date3.dropna()
df_date3.tail()

returns3 = df_date3["Return"]

# ----- Empirical VaR -----
returns_sorted3 = np.sort(returns3)
N3 = len(returns_sorted3)
index_empirical3 = int(np.floor(alpha * (N3 - 1)))
VaR_empirical3 = returns_sorted3[index_empirical3]
print(f"Empirical VaR at {alpha*100}%: {VaR_empirical3:.4f}")

# ----- Parametric VaR -----
mu3 = returns3.mean()
sigma3 = returns3.std(ddof=1)
z3 = norm.ppf(alpha)
VaR_parametric3 = mu3 + sigma3 * z3
print(f"Parametric Gaussian VaR at {alpha*100}%: {VaR_parametric3:.4f}")

# ----- KDE VaR -----
kde3 = gaussian_kde(returns3)
x_grid3 = np.linspace(returns3.min(), returns3.max(), 10000)
cdf3 = np.cumsum(kde3(x_grid3))
cdf3 /= cdf3[-1]
VaR_kernel3 = x_grid3[np.searchsorted(cdf3, alpha)]
print(f"Non-parametric KDE VaR at {alpha*100}%: {VaR_kernel3:.4f}")

# ----- Validation -----
methods3 = {
    "Empirical": VaR_empirical3,
    "Parametric": VaR_parametric3,
    "Kernel": VaR_kernel3
}

results3 = []
for name, VaR in methods3.items():
    exceptions3 = int(np.sum(returns3 <= VaR))
    proportion3 = exceptions3 / N3
    results3.append([name, exceptions3, proportion3])

# ----- Tableau final -----
table3 = pd.DataFrame(results3, columns=["Method", "Exceptions", "Proportion"])
print(table3.to_string(index=False))
```
